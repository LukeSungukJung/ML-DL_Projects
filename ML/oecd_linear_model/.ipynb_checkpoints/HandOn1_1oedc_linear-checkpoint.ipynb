```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn.linear_model
import warnings
warnings.filterwarnings('ignore')
```

```python

oecd_bli = pd.read_csv('2015_oecd_bli.csv',thousands=',')
oecd_bli.head()
```

```python
oecd_bli.isnull().sum()
```

```python
oecd_bli.describe()
```

```python
life_satisfaction = oecd_bli['  Life satisfaction']
earnings = oecd_bli['  Personal earnings']
```

```python
y=np.c_[life_satisfaction]
x=np.c_[earnings]
```

```python
plt.scatter(x,y,marker='*',c='g')
plt.grid(True)
plt.show()
```

```python
#y= x*w0+ w1
```

```python
model = sklearn.linear_model.LinearRegression()
```

```python
model.fit(x,y)
```

```python
import scipy as sp

cor_abs = abs(oecd_bli.corr(method='spearman'))
cor_cols = cor_abs.nlargest(n=24, columns="  Life satisfaction").index
```

```python
cor = np.array(sp.stats.spearmanr(oecd_bli[cor_cols].values))[0]
plt.figure(figsize=(15,15))
sns.set(font_scale=1)
sns.heatmap(cor,fmt='.2f', annot=True, square=True, annot_kws={'size': 8}, xticklabels=cor_cols.values,yticklabels=cor_cols.values)
```
